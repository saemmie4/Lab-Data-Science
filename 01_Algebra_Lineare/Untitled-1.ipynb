```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib notebook
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
img=mpimg.imread('bolla1.jpg')
```

```python
img.shape
```

```python
fig, ax = plt.subplots()

plt.imshow(img[:,:,:])
plt.show()
```

```python
X = rgb2gray(img/255.)

fig, ax = plt.subplots()

plt.imshow(X, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

```python
X.shape
```

```python
U, S, Vh = np.linalg.svd(X, full_matrices=True)
```

```python
fig, ax = plt.subplots()

ax.set_xlabel('index')
ax.set_ylabel('singular values')

plt.semilogy()
ax.plot(S,lw=3)
```

```python
CumEn = np.cumsum(S)/np.sum(S)
```

```python
fig, ax = plt.subplots()

ax.set_xlabel('index')
ax.set_ylabel('Cumulative Energy')

ax.set_ylim([0,1.1])

ax.plot(CumEn,lw=3)
```

```python
r = 100

U_tilde  = U[:,:r]
S_tilde  = S[:r]
Vh_tilde = Vh[:r,:]

X_tilde = U_tilde @ np.diag(S_tilde) @ Vh_tilde
```

```python
fig, ax = plt.subplots()

plt.imshow(X_tilde, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```
